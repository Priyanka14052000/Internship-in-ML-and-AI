```
import pandas as pd
from google.colab import files

file = files.upload()
X_train = pd.read_csv('xtrain.csv')
Y_train = pd.read_csv('ytrain.csv')
X_test = pd.read_csv('xtest.csv')
Y_test = pd.read_csv('ytest.csv')
```

This neural network can be used to predict future values

```
from keras.models import Sequential
from keras.layers import Dense

classifier = Sequential() # initialising the ANN

classifier.add(Dense(units = 16, activation = 'relu', input_dim = 30))
classifier.add(Dense(units = 8, activation = 'relu'))
classifier.add(Dense(units = 6, activation = 'relu'))
classifier.add(Dense(units = 1, activation = 'sigmoid'))
```

Once the neural network is defined, optimiser and loss function is specified

```
classifier.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy')
```

Train the Neural Network using classifier.fit -- for set of X, how the Y should look like. The the Neural Networks will spot patterns in the data

```
classifier.fit(X_train, Y_train, batch_size = 1, epochs = 100)
```

```
Y_pred = classifier.predict(X_test)
Y_pred = [1 if y >= 0.5 else 0 for y in Y_pred]
```

```
total = 0
correct = 0
wrong = 0

for i in range(len(Y_pred)):
  total = total + 1
  if(Y_test.iat[i,0] == Y_pred[i]):
    correct = correct + 1
  else:
    wrong = wrong + 1
    
print('Total ' + str(total))
print('Correct ' + str(correct))
print('Wrong ' + str(wrong))
```
