## Task 1 : Draw a pie plot based on number of wins for each team

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import timeit

class winTeam():
    def generatePieChart(self, matchdata):
        start = timeit.default_timer()
        
        # collecting the data
        df = pd.read_csv(matchdata)
        
        wins, winner = np.unique(list(df['match_winner'].values), return_counts = True)
        match_wins, no_wins = wins[:13], winner[:13]
    
        
        # generating pie chart with the collected data
        label = match_wins
        data = no_wins

        explode = [0.1 if i == max(no_wins) else 0 for i in no_wins]
        
        fig1, ax1 = plt.subplots()
        ax1.pie(data, labels=label, explode = explode, autopct='%1.2f%%',
                shadow = False, startangle=90)
        ax1.axis('equal') 
        plt.show()
        
        return timeit.default_timer() - start

match = winTeam()
match.generatePieChart('Match.csv')
```

## Task 2 : Draw a bar graph for number of matches won (i,e.consider win if the team wins both the toss and match)  for the team “Mumbai Indians” for all years

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import timeit

class winMumbaiIndians():
    def generateBarGraph(self, matchdata):
        start = timeit.default_timer()
        
        # collecting the data
        df = pd.read_csv(matchdata)
        
        match_winner = df['match_winner'].values
        toss_winner = df['Toss_Winner'].values
        
        years = np.unique(df['Season_Year'].values)

        win_2008 = list(df.iat[i,10] for i in range(0, len(df)) if df.iat[i,5] == 2008 if df.iat[i,9] == 'Mumbai Indians' and df.iat[i,10] == 'Mumbai Indians')
        win_2009 = list(df.iat[i,10] for i in range(0, len(df)) if df.iat[i,5] == 2009 if df.iat[i,9] == 'Mumbai Indians' and df.iat[i,10] == 'Mumbai Indians')
        win_2010 = list(df.iat[i,10] for i in range(0, len(df)) if df.iat[i,5] == 2010 if df.iat[i,9] == 'Mumbai Indians' and df.iat[i,10] == 'Mumbai Indians')
        win_2011 = list(df.iat[i,10] for i in range(0, len(df)) if df.iat[i,5] == 2011 if df.iat[i,9] == 'Mumbai Indians' and df.iat[i,10] == 'Mumbai Indians')
        win_2012 = list(df.iat[i,10] for i in range(0, len(df)) if df.iat[i,5] == 2012 if df.iat[i,9] == 'Mumbai Indians' and df.iat[i,10] == 'Mumbai Indians')
        win_2013 = list(df.iat[i,10] for i in range(0, len(df)) if df.iat[i,5] == 2013 if df.iat[i,9] == 'Mumbai Indians' and df.iat[i,10] == 'Mumbai Indians')
        win_2014 = list(df.iat[i,10] for i in range(0, len(df)) if df.iat[i,5] == 2014 if df.iat[i,9] == 'Mumbai Indians' and df.iat[i,10] == 'Mumbai Indians')
        win_2015 = list(df.iat[i,10] for i in range(0, len(df)) if df.iat[i,5] == 2015 if df.iat[i,9] == 'Mumbai Indians' and df.iat[i,10] == 'Mumbai Indians')
        win_2016 = list(df.iat[i,10] for i in range(0, len(df)) if df.iat[i,5] == 2016 if df.iat[i,9] == 'Mumbai Indians' and df.iat[i,10] == 'Mumbai Indians')
        win_2017 = list(df.iat[i,10] for i in range(0, len(df)) if df.iat[i,5] == 2017 if df.iat[i,9] == 'Mumbai Indians' and df.iat[i,10] == 'Mumbai Indians')
        
        no_wins = [len(win_2008), len(win_2009), len(win_2010), len(win_2011), len(win_2012), len(win_2013), len(win_2014),
                  len(win_2015), len(win_2016), len(win_2017),]
        
        # generating bar graph with the collected data
        width = 1/2.5
        plt.title("Mumbai Indians' wins over the years(2008 - 2017)")
        plt.bar(years, no_wins, width, color = 'gray')
        
        return timeit.default_timer() - start

match = winMumbaiIndians()
match.generateBarGraph('Match.csv')
```
