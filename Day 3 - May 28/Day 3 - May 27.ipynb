# Task 1 :
### Construct a Dictionary with keys as Genre names and values as frequency percentage  for each genre.

Genre | App Count % 
--- | --- 
Games | ... 
Productivity | ... 
Social Networking | ...
Health Care | ...
etc | ...

#  

```python
import timeit
import numpy as np
import pandas as pd

def generateAppData(AppData):
    start = timeit.default_timer()
    genre, category, freq, percent, j = [], [], [], {}, 0
    df = pd.read_csv(AppData)
    
    for i in df.values:
        genre.append(i[12])
        
    category, freq = np.unique(genre, return_counts = True)
    per_freq = list(map(lambda i : float('{:.2f}'.format((i/sum(freq))*100)), freq))
    
    for i in category:
        percent[i] = per_freq[j]
        j += 1
        
    print('Percentage of each genre: ', end = '\n\n')
    for i in percent.items():
        print(i)
            
    return timeit.default_timer() - start
generateAppData('AppleStore1.csv')
```
